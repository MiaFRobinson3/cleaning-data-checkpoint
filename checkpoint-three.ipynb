## Missing Data

Test your dataset for missing data and handle it as needed. Make notes in the form of code comments as to your thought process.

```python
dataframe.isna()
dataframe.isnull()
missing_values = dataframe.isnull()

print(missing_values)
#I am looking for null values. At first glance it would appear there are not a lot of null values. 
```

```python
for col in dataframe.columns:
    pct_missing = np.mean(dataframe[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))

#I am looking to find the percentage of null or missing values. Again, it does not appear there are many null values, but there are some. 
```

```python
dataframe.isnull().sum()
```

```python
dataframe_cleaned = dataframe.dropna()
print(dataframe_cleaned)

dataframe_cleaned.isna()
dataframe_cleaned.isnull()
#Since there are so few null values, I dropped them from a cleaned version of my dataframe. I then checked to see that they were dropped. 
```
